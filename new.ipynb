```python
import requests
import warnings
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from concurrent.futures import ThreadPoolExecutor
```

```python
url = 'https://www.geeksforgeeks.org/python-projects-beginner-to-advanced/?ref=footer'
response = requests.get(url)
html_content = response.text 
soup = BeautifulSoup(html_content,'html.parser')
```

```python
list1 = []
links = []
```

```python
Projects_Beginners = soup.find('div',{'class':'text'})
doc_content = [element.text.strip() for element in Projects_Beginners.find_all(['li'])]
doc_content
```

```python
urls = [urljoin(url, link.get('href')) for link in Projects_Beginners.find_all('a')]
urls = urls[1:]  # يتم حذف العنصر الأول
urls
```

```python
list1.extend(doc_content)
links.extend(urls)
```

```python
print(len(links))
```

```python
import pandas as pd
data = {'ProjectName':list1,'Links':links}
df = pd.DataFrame(data)
df.to_excel('pythonproject.xlsx', index=False)
```
